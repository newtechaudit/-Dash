```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# считываем и сохраняем в df исходный файл
data = pd.read_csv('games.csv')
```

```python
data
```

```python
data.info()
```

В данном случае нам не принципиально кол-во записей для создания примера интерактивных графиков в dash (python). Поэтому просто удалим все записи с пропусками.

```python
# удаление строк с NaN значениями
data = data.dropna()
data
```

```python
data.info()
```

Удалим строки с tbd значениями.  
PS - tbd это аббревиатура to be decided / to be determined

```python
# удаление строк с tbd значениями
data = data[data['User_Score'] != 'tbd']
data
```

```python
data.info()
```

Тип данных у года - float, преобразуем в int

```python
data['Year_of_Release']
```

```python
data['Year_of_Release'] = data['Year_of_Release'].astype(int) # перевод из float в тип int
```

```python
data.info()
```

```python
data.sample(10)
```

Оставим только записи с годом >= 2000

```python
data = data[data['Year_of_Release'] >= 2000] 
```

```python
data.info()
```

```python
# удаление дубликатов
data = data.drop_duplicates()
```

```python
data['User_Score'] = data['User_Score'].astype(float) # переводим оценку пользователей в float
```

```python
data['Critic_Score'] = data['Critic_Score'].astype(int) # переводим оценку критиков в int
```

```python
data.info()
```

```python
data.sample(30)
```

```python
# сохранение измененный файл в новый файл Norm_games.h5
data.to_hdf('Norm_games.h5', key='data', mode='a')
```

```python
# проверка корректности записи файла
new_hdf = pd.read_hdf('Norm_games.h5')
new_hdf
```
